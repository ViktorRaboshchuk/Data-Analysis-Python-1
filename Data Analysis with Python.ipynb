# Sorting

```python
# We can sorn dataframe in the same way we sorting Excel file
# Let's sort dataframe by column 'Total day minutes' with descending order
df.sort_values(by='Total day minutes', ascending = False).head()
```

```python
# We also can sort dataframes by multiple columns
df.sort_values(by=['State', 'Total day charge'],ascending=[True, False]).head()
```

# Indexing and retrieving data

```python
# Here we can see proportion of churned users in our dataframe = 14.5%
df['Churn'].mean()
```

```python
# Here we take only churned users (df['Churn']==1) and extracting average values by every column
df[df['Churn']==1].mean()
```

```python
# But if we want to see average value of one column we have to add its name
df[df['Churn']==1]['Total day minutes'].mean()
```

```python
# Writing some logical statements 
# So we are searching for loyal users calls duration that don't have international plan
df[df['Churn']==0 & (df['International plan']=='No')]['Total intl minutes'].max()
```

```python
# Here you can see different ways to index dataframes. 
# loc()  method use indexing by name
df.loc[0:5, 'State':'Area code']
```

```python
# iloc() method use indexing by number
df.iloc[0:5, 0:3]
```

# Applying functions to Cells, Columns and Rows

```python
# To apply function to each dataframe's column use apply() function
df.apply(np.max)
```

```python
# If you want to see each state starting by 'R' use lambda function
df[df['State'].apply(lambda state: state[0] == 'R')].head()
```

```python
# If you want to replace some values in the dataframe with new one use replase method
d = {'No' : False, 'Yes' : True}
df = df.replace({'Voice mail plan': d})
df.head()
```

# Grouping

```python
# Standart way of grouping in pandas
# df.groupby(by=grouping_columns)[columns_to_show].function()
```

```python
# Here we use grouping dataframe by 'Churm' column , getting only 3 column, and for this 3 column applying describe() method
To_show = ['Total day calls','Total eve calls','Total night calls']
df.groupby(['Churn'])[To_show].describe(percentiles=[])
```

```python
df.groupby(['Churn'])[To_show].agg([np.mean,np.std,np.min,np.max])
```

# Summary tables

```python
# You should use contingency table using the crosstab method to see some observations in context of two variables
```

```python
pd.crosstab(df['Churn'], df['International plan'])
```

```python
pd.crosstab(df['Churn'], df['Voice mail plan'], normalize = True)
```

```python
# Examples of pivot tables, that we also have in Excel
```

```python
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'],
               ['Area code'], aggfunc='mean')
```

# DataFrame transformations

```python
# One way to add new column to existing dataframe
df['Total charge'] = df['Total day charge'] + df['Total eve charge'] + \
                     df['Total night charge'] + df['Total intl charge']
df.head()
```

```python
# Deleting column
df.drop(['Total charge'],axis =1,inplace=True)
df.head()
```
